```
import os
```

```
sktch_fol_path='/content/gdrive/My Drive/SBIR Dataset/sketches'
```

```
os.listdir(sktch_fol_path)
#!ls "/content/gdrive/My Drive"
```

```
class_names=os.listdir(sktch_fol_path)#.__next__()[1]
```

```
class_names
```

```
class_paths=[]
for each_class in class_names:
  #print(os.path.join(sktch_fol_path,each_class))
  class_paths.append(os.path.join(sktch_fol_path,each_class))
```

```
os.listdir(class_paths[0])
```

```
#import matplotlib.pyplot as plt
#from google.colab.patches import cv2
#from PIL import Image
import cv2
#os.listdir(class_paths[0])
for class_name in class_names:
  curr_class_folder=os.path.join(sktch_fol_path,class_name)
  #print(curr_class_folder)
  imgs=os.listdir(curr_class_folder)
  split1=int(0.8*len(imgs))
  split2=int(0.9*len(imgs))
  training_imgs=imgs[:split1]
  val_imgs=imgs[split1:split2]
  test_imgs=imgs[split2:]
  os.mkdir(os.path.join('/content/gdrive/My Drive/Splitted_SBIR_Data/Sketches/Training',class_name))
  training_target_path=os.path.join('/content/gdrive/My Drive/Splitted_SBIR_Data/Sketches/Training',class_name)
  #print(imgs)
  for img_name in training_imgs:
    training_img_path=os.path.join(curr_class_folder,img_name)
    i=cv2.imread(training_img_path)
    cv2.imwrite(os.path.join(training_target_path,img_name),i)
  #break
  os.mkdir(os.path.join('/content/gdrive/My Drive/Splitted_SBIR_Data/Sketches/Testing',class_name))
  test_target_path=os.path.join('/content/gdrive/My Drive/Splitted_SBIR_Data/Sketches/Testing',class_name)
  #print(imgs)
  for img_name in test_imgs:
    training_img_path=os.path.join(curr_class_folder,img_name)
    i=cv2.imread(training_img_path)
    cv2.imwrite(os.path.join(test_target_path,img_name),i)
  #break
  os.mkdir(os.path.join('/content/gdrive/My Drive/Splitted_SBIR_Data/Sketches/Validation',class_name))
  val_target_path=os.path.join('/content/gdrive/My Drive/Splitted_SBIR_Data/Sketches/Validation',class_name)
  #print(imgs)
  for img_name in val_imgs:
    training_img_path=os.path.join(curr_class_folder,img_name)
    i=cv2.imread(training_img_path)
    cv2.imwrite(os.path.join(val_target_path,img_name),i)
  #break
```
