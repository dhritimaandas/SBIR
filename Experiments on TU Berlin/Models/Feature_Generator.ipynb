## **Image_Feature_Generator**

```
feature_gen=load_model('/content/gdrive/My Drive/Best_Model/Image/Img_Feature_generator.h5',compile=True)
```

```
feature_gen.summary()
```

```
img_dir='/content/gdrive/My Drive/Splitted_SBIR_Data/Images_Balanced/Training'
```

```
import os
labels=os.listdir(img_dir)
```

```
from keras.applications.vgg16 import preprocess_input
```

```
import cv2
import numpy as np
for label in labels:
  x=[]
  label_dir=os.path.join(img_dir,label)
  img_names=os.listdir(label_dir)
  os.mkdir(os.path.join('/content/gdrive/My Drive/Feature_Vectors/Image/Training/',label))
  for image in img_names:
    img_path=os.path.join(label_dir,image)
    i=cv2.imread(img_path)
    i=cv2.resize(i,(224,224,))
    i=np.expand_dims(i, axis=0)
    new_inp=preprocess_input(i)
    final_fea=np.squeeze(feature_gen.predict(new_inp))
    np.save(os.path.join('/content/gdrive/My Drive/Feature_Vectors/Image/Training/',label,os.path.splitext(image)[0]), final_fea)
```
